# Step 2

```python
#Combining the 4 sepearter q#_issr into 1 dataframe called df_issuers creating a new column called Quarter
df_issuers = pd.concat([q1_issr, q2_issr, q3_issr, q4_issr], keys = ['Q1', 'Q2', 'Q3', 'Q4'], names = ['Quarter', None]).reset_index(level = 0)
df_issuers = df_issuers.reset_index(drop = True)
df_issuers
```

```python
#Combining the 4 sepearter q#_disc into 1 dataframe called df_disc
df_disc = pd.concat([q1_disc, q2_disc, q3_disc, q4_disc])
df_disc = df_disc.reset_index(drop = True)
df_disc
```

```python
#dropping unneccesary columns in df_issuers
df_issuers.drop(['ISAMENDMENT',
                 'PROGRESSUPDATE',
                 'NATUREOFAMENDMENT',
                 'LEGALSTATUSOTHERDESC',
                 'STREET1',
                 'STREET2',
                 'ZIPCODE', 
                 'ISSUERWEBSITE', 
                 'COMMISSIONCIK', 
                 'COMMISSIONFILENUMBER', 
                 'CRDNUMBER' ], axis = 1, inplace = True)
df_issuers.head()
```

```python
#dropping unneccesary columns in df_disc
df_disc.drop(['COMPENSATIONAMOUNT',
              'FINANCIALINTEREST',
              'SECURITYOFFEREDTYPE', 
              'SECURITYOFFEREDOTHERDESC', 
              'PRICEDETERMINATIONMETHOD',
              'OVERSUBSCRIPTIONALLOCATIONTYPE',
              'DESCOVERSUBSCRIPTION',
              'MAXIMUMOFFERINGAMOUNT',
              'CASHEQUIMOSTRECENTFISCALYEAR',
              'CASHEQUIPRIORFISCALYEAR',
              'ACTRECEIVEDRECENTFISCALYEAR',
              'ACTRECEIVEDPRIORFISCALYEAR',
              'SHORTTERMDEBTMRECENTFISCALYEAR',
              'SHORTTERMDEBTPRIORFISCALYEAR',
              'COSTGOODSSOLDRECENTFISCALYEAR',
              'COSTGOODSSOLDPRIORFISCALYEAR',
              'TAXPAIDMOSTRECENTFISCALYEAR',
              'TAXPAIDPRIORFISCALYEAR'], axis = 1, inplace = True)
df_disc.head()
```

```python
#Merging ds_issuers and df_disc into on ID ACCENSION_NUMBER into a new dataframe called df_crowd
df_crowd = pd.merge(df_issuers, df_disc, on = 'ACCESSION_NUMBER', how = 'outer')
df_crowd
```

```python
#Changing the column names in df_crowd to lowercase
df_crowd.columns = df_crowd.columns.str.lower()
df_crowd.head()
```

```python
#renaming columns to add in _ to make data easier to read
df_crowd = df_crowd.rename(columns = {'nameofissuer' : 'issuer_name',
                                      'legalstatusform' : 'form_of_organization', 
                                      'jurisdictionorganization' : 'registered_jurisdiction', 
                                      'dateincorporation' : 'incorporation_date',
                                      'stateorcountry' : 'state_or_country',
                                      'companyname' : 'company_name',
                                      'noofsecurityoffered' : 'securities_offered',
                                      'offeringamount' : 'offering_amount', 
                                      'oversubscriptionaccepted' : 'oversubscription_accepted', 
                                      'deadlinedate' : 'deadline_date',
                                      'currentemployees' : 'current_employees',})
df_crowd.head()
```

```python
#Seeing where the null values are in each column
df_crowd.isnull().sum()
```

```python
df_crowd = df_crowd.dropna()
```

```python
df_crowd.isnull().sum()
```

```python
#checking column information
df_crowd.info()
```

```python
#changing incorporation_date and deadline_date to datetime object
df_crowd.incorporation_date = pd.to_datetime(df_crowd.incorporation_date)
df_crowd.deadline_date = pd.to_datetime(df_crowd.deadline_date)
df_crowd.info()
```

# Step 3

```python
#Revenue Growth
pd.options.mode.chained_assignment = None  # default='warn'
# adding rev growth col
df_crowd.loc[:, "revgrowth"] = df_crowd.revenuemostrecentfiscalyear - df_crowd.revenuepriorfiscalyear
```

```python
#showing top 10 companies with highest revenue growth
df_crowd[['issuer_name','revenuemostrecentfiscalyear','revenuepriorfiscalyear', 'revgrowth'
         ]].nlargest(10,'revgrowth')
```

```python
# Revenue growth is an important metric to look at. Companies should be expanding sales and try not to remain stagnate.
# high revenue growth indicates a profitable business model and suggests that the business can continue to grow.
#Future investors can look at this metric to determine if a business is expanding which will let them know their 
#investment is more likely to pay off.

#Point n largest was a new python function that was looked up and learned.
```
