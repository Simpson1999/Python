```python
import pandas as pd

import datetime

import numpy as np

import matplotlib.pyplot as plt

from pandas.plotting import scatter_matrix

!pip install yfinance

import yfinance as yf

%matplotlib inline
```

```python
start = "2021-04-25"
end = '2022-04-25'
tsla = yf.download('TSLA',start,end)
msft = yf.download('MSFT',start,end)
aapl = yf.download('AAPL',start,end)
```

```python
#1 Year
tsla['Close'].plot(label = 'TSLA', figsize = (15,7))
msft['Close'].plot(label = "MSFT")
aapl['Close'].plot(label = 'AAPL')
plt.title('1 Year Stock Prices of Tesla, Microsoft and Apple')
plt.legend()
```

```python
tickers = ['msft', 'tsla', 'aapl']
stock_data = yf.download(tickers, start="2021-04-25", end="2022-04-25")
```

```python
#1 Year Correlation
returns = np.log(stock_data['Adj Close']/ stock_data['Adj Close'].shift(1))
returns.dropna(inplace=True)
returns.plot(subplots=True, figsize=(14,6))
```

```python
returns.corr()
```

Correlation is a statistical term that describes how closely two variables are connected in a linear fashion. It's a typical way of explaining simple interactions without stating a cause and effect relationship. The change in frequencies affect the correlation heavily, as there is a different amount of days and stock prices included in those time horizons. For example you can view the differences between Microsoft and Tesla for the 3-month and 1-month horizons at .61 for the 3-month and .71 for the 1 month. They nearly have a .1 difference for those two timelines and that presents a massive change in how correlated those two are.
